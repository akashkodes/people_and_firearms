```python
df =  pd.read_csv('data875.csv', header=0, keep_default_na=False).values.tolist()
```

```python
df = pd.DataFrame(df)
df.columns = ['sr_no', 'year', 'state', 'permits', 'handguns', 'long_guns', 'bg_check', 'avg_age', 'avg_education', 'avg_income', 'per_female', 'per_hispanic', 'per_white', 'per_black', 'per_native', 'per_asian']
print(df)
```

```python
cols = df.columns.drop('state')

df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df['total_arms'] = df['handguns'] + df['long_guns']
```
